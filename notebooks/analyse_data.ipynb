```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python

def analyze_data_quality(df):
    # Class distribution
    print("Class Distribution:")
    print(df['Diagnosis'].value_counts(normalize=True))

    # Missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Feature correlations
    plt.figure(figsize=(10, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title('Feature Correlations')
    plt.show()

    # Feature distributions
    for column in df.select_dtypes(include=[np.number]).columns:
        if column != 'Diagnosis':
            plt.figure(figsize=(10, 4))

            # Distribution plot
            plt.subplot(121)
            sns.histplot(data=df, x=column, hue='Diagnosis', bins=30)
            plt.title(f'{column} Distribution by Class')

            # Box plot
            plt.subplot(122)
            sns.boxplot(data=df, x='Diagnosis', y=column)
            plt.title(f'{column} Box Plot by Class')

            plt.tight_layout()
            plt.show()

    # Basic statistics
    print("\nFeature Statistics:")
    print(df.describe())
```

```python

# Load and analyze data
train_data = pd.read_csv('../data/processed_data/train_data.csv')
analyze_data_quality(train_data)
```
